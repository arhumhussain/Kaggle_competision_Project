#### Imputing categorical NAN with mode having missing values less than 6% and assign "Missing" where Missing values is Greter than 6%

```python
 for i in cate_nan:
        if copy[i].isnull().mean()*100<=6:
            copy[i]=copy[i].fillna(copy[i].mode()[0])
        else:
            copy[i]=np.where(copy[i].isnull(),'Missing',copy[i])
```

### check distribution before and after Imputing NaN

```python
def dist(original,df,i):
    fig=plt.figure(figsize=(12,3))
    ax=fig.add_subplot(121)
    original[original[i]==original[i].mode()[0]]['SalePrice'].plot(kind='kde',ax=ax)
    df[df[i]==df[i].mode()[0]]['SalePrice'].plot(kind='kde',ax=ax,color='red')
    lines,labels=ax.get_legend_handles_labels()
    labels=['Not Null','Mode Imputed']
    ax.legend(lines,labels,loc='best')
    plt.title('imputed {}'.format(i))
    
    #fig=plt.figure()
    ax=fig.add_subplot(122)
    original[original[i]==original[i].mode()[0]]['SalePrice'].plot(kind='kde',ax=ax)
    original[original[i].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='red')
    lines,labels=ax.get_legend_handles_labels()
    labels=['Not null','Null']
    ax.legend(lines,labels,loc='best')
    plt.title("Original {}".format(i))
    plt.show()
```

# Outliers

```python
def outlier(data,feature): # Function for Outlier Detection
    IQR=data[feature].quantile(0.75)-data[feature].quantile(0.25)
    up=data[feature].quantile(0.75)+(IQR*1.5)
    low=data[feature].quantile(0.25)-(IQR*1.5)
    out=data[(data[feature]>up) | (data[feature]<low)]
    return len(out)*100/len(data)
```

```python
for i in numerical_features:
    if 0 in copy[i].unique():
        pass
    else:
        print("outliers in ",i,"is",round(outlier(copy,i),2),'%')
```

### Caping Outliers

```python
copy1=copy.copy()
for i in numerical_features:
    if outlier(copy1,i)>0 and 0 not in copy1[i].unique():
        IQR=copy1[i].quantile(0.75)-copy[i].quantile(0.25)
        up=copy1[i].quantile(0.75)+(IQR*1.5)
        low=copy1[i].quantile(0.25)-(IQR*1.5)
        copy1[i]=np.where((copy1[i]>up),up,np.where(copy1[i]<low,low,copy1[i]))
        plt.figure(figsize=(12,3.5))
        plt.subplot(1,2,1)
        plt.title("after")
        plt.xlabel(i)
        sns.boxplot(copy1[i])
        plt.subplot(1,2,2)
        plt.title('Before')
        plt.xlabel(i)
        sns.boxplot(copy[i],color='red')
        plt.show()
    else:
        pass
```

# RARE Categories in features
if the category exist less than 1% in column assign them RARE_LABEL

```python
copy2=copy1.copy()
for i in [i for i in copy2.columns if copy2[i].dtype=="O"]:
    rar=copy2.groupby(i)['SalePrice'].count()/len(copy2[i])
    rar_f=rar[rar>0.01].index
    copy2[i]=np.where(copy2[i].isin(rar_f),copy2[i],'Rare_label')
    
```
